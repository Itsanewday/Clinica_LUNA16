```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pd.read_excel('F:/Prac_Clinicas/LIDC-IDRI/Nodule_Binarization.xlsx', index_col=0)
Nodules = np.array(df)
```

```python
Images_1 = []
Images_0 = []

for i in range(len(Nodules)):
    pickle_in = open('F:\Prac_Clinicas\Images\Resampled_pickle_120\Patient_{}.pickle'.format(i+1), 'rb')
    Patient = pickle.load(pickle_in)
    if Nodules[i] == 1:
        Images_1.append(Patient)
        print('Class 1: {}'.format(len(Images_1)))
    
    else:
        Images_0.append(Patient)
        print('Class 0: {}'.format(len(Images_0)))
        
with open ('F:\Prac_Clinicas\Images\Final_Pickle\Images_120px_1.pickle', 'wb') as f:
    pickle.dump(Images_1, f)
    
with open ('F:\Prac_Clinicas\Images\Final_Pickle\Images_120px_0.pickle', 'wb') as f:
    pickle.dump(Images_0, f)
        
```
