### Patients

```python
# Some constants 
INPUT_FOLDER = 'E:/Prac_Clinicas/Images/LIDC-IDRI/'
patients = os.listdir(INPUT_FOLDER)
patients.sort()
```

### Segmented Lungs

```python
for num, patient in enumerate(patients):
    Segmented_Lungs = []
    print ('Patient: {}'.format(num+1))
    patient_number = lung.load_scan(INPUT_FOLDER + patient)
    patient_pixels = lung.get_pixels_hu(patient_number)
    segmented_patient = []
    for slice in patient_pixels:
        segmented = lung.seperate_lungs(slice)
        segmented_patient.append(segmented)
    Segmented_Lungs.append(segmented_patient)
    with open ('E:\Prac_Clinicas\Images\Pickle\Seg_Pat_{}.pickle'.format(num+1), 'wb') as f:
        pickle.dump(Segmented_Lungs, f)
```
