# Segmented Images (Pickle)

### Libraries

```python
%matplotlib notebook

import numpy as np # linear algebra
import os
import matplotlib.pyplot as plt
import pickle
import Lung_Segmentation as lung
```

### Patients

```python
# Some constants 
INPUT_FOLDER = 'F:/Prac_Clinicas/Images/LIDC-IDRI/'
patients = os.listdir(INPUT_FOLDER)
patients.sort()
```

```python
for num, patient in enumerate(patients[294:400]):
    print('\n'+patient)
    Patient_Scan = lung.load_scan(INPUT_FOLDER + patient)
    pickle_in = open('F:\Prac_Clinicas\Images\Pickle\Seg_Pat_{}.pickle'.format(num+295), 'rb')
    Segmented_Lung = pickle.load(pickle_in)
    Segmented_patient = np.array(Segmented_Lung)
    Resampled_Lung = lung.resample(Segmented_patient[0], Patient_Scan)
    with open ('F:\Prac_Clinicas\Images\Resampled_pickle_120\Patient_{}.pickle'.format(num+295), 'wb') as f:
        pickle.dump(Resampled_Lung, f)
```
